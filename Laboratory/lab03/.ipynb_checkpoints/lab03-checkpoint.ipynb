```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math

x0 = 38000
y0 = 29000
t0 = 0
tmax = 1

a = 0.41
b = 0.76
c = 0.59
h = 0.63

a2 = 0.37
b2 = 0.76
c2 = 0.32
h2 = 0.61

def P(t):
    p= math.fabs(math.sin(t+3))
    return p

def Q(t):
    q= math.fabs(math.cos(t+2))
    return q


def P2(t):
    p= math.fabs(math.sin(6*t))
    return p

def Q2(t):
    q= math.fabs(math.cos(7*t))
    return q

def f(y, t):
    y1, y2 = y
    return [-a*y1 - b*y2 + P(t), -c*y1 - h*y2 + Q(t) ]

def f2(y, t):
    y1, y2 = y
    return [-a2*y1 - b2*y2 + P2(t), -c2*y1*y2 - h2*y2 + Q2(t) ]

t = np.linspace( 0, tmax, num = 100)
y0 = [x0, y0]
w1 = odeint(f, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
fig = plt.figure(facecolor='white')
plt.plot(t, y11, t, y21, linewidth=2)
plt.ylabel("x, y")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('03.png', dpi = 600)

w1 = odeint(f2, y0, t)
y12 = w1[:,0]
y22 = w1[:,1]
fig2 = plt.figure(facecolor='white')
plt.plot(t, y12, t, y22, linewidth=2)
plt.ylabel("x, y")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig2.savefig('04.png', dpi = 600)
```
